```python
## Task 2: Cleaning the Data
## Subtask 2.1: Inspect Null values¶
```

```python
 null_counts = df.isnull().sum()

print("Column-wise Null Counts:")
print(null_counts)
```

```python
# code for row-wise null count
null_counts = df.isnull().sum(axis=1)
print("Row-wise Null Counts:")
print(null_counts)
```

```python
#code for column-wise null percentages 
total_rows = len(df)
null_counts = df.isnull().sum()
null_percentages = (null_counts / total_rows) * 100
null_percentages = null_percentages.round(2)

print("Column-wise Null Percentages:")
print(null_percentages)
```

```python
# Subtask 2.2: Drop unecessary columns
df.drop(['color','director_facebook_likes', 'actor_1_facebook_likes', 'actor_2_facebook_likes', 'actor_3_facebook_likes', 'actor_2_name', 'cast_total_facebook_likes', 'actor_3_name', 'duration', 'facenumber_in_poster', 'content_rating', 'country', 'movie_imdb_link', 'aspect_ratio', 'plot_keywords'],axis=1)
```

```python
df
```

```python
# Subtask 2.3: Drop unecessary rows using columns with high Null percentages
null_percentages = (df.isnull().sum() / len(df)) * 100
columns_to_drop = null_percentages[null_percentages > 5].index
df = df.dropna(subset=columns_to_drop)
print("DataFrame after dropping rows with null values for columns with >5% null values:")
print(df)
```

```python
# Subtask 2.4: Drop unecessary rows
nan_counts = df.isnull().sum(axis=1)
df = df[nan_counts <= 5]
print("DataFrame after removing rows with more than five NaN values:")
print(df)
```

```python
# Subtask 2.5: Fill NaN values
df['language'].fillna('English', inplace=True)
print("DataFrame after replacing missing values in 'language' column with 'English':")
print(df)
```

```python
# Subtask 2.6: Check the number of retained rows

df['language'].fillna('English', inplace=True)
num_rows_before = len(df)
nan_counts = df.isnull().sum(axis=1)
df = df[nan_counts <= 5]
num_rows_after_nan_removal = len(df)
print("Number of rows before: ", num_rows_before)
print("Number of rows after removing rows with more than five NaN values: ", num_rows_after_nan_removal)
print("Percentage of rows retained: {:.2f}%".format((num_rows_after_nan_removal / num_rows_before) * 100))
```

```python
# Task 3: Data Analysis
# Subtask 3.1: Change the unit of columns
```

```python
df['budget'] = df['budget'] / 1000000  # Convert budget column to million dollars
df['gross'] = df['gross'] / 1000000    # Convert gross column to million dollars

print("DataFrame after converting budget and gross columns to million dollars:")
print(df)
```

```python
# Subtask 3.2: Find the movies with highest profit
## Create a new column called profit which contains the difference of the two columns: gross and budget.
df['profit'] = df['gross'] - df['budget']

print("DataFrame with the new 'profit' column:")
print(df)
```

```python
# Sort the dataframe using the profit column as reference
df_sorted = df.sort_values(by='profit')

print("DataFrame sorted by the 'profit' column:")
print(df_sorted)
```

```python
# Extract the top ten profiting movies in descending order and store them in a new dataframe - top10

df_sorted = df.sort_values(by='profit', ascending=False)
top10 = df_sorted.head(10)
print("Top ten profiting movies:")
print(top10)
```

```python
## Subtask 3.3: Drop duplicate values

df_unique = df.drop_duplicates()
df_sorted_unique = df_unique.sort_values(by='profit', ascending=False)
top10_unique = df_sorted_unique.head(10)
print("Top ten profiting movies after dropping duplicates:")
print(top10_unique)
```

```python
# Subtask 3.4: Find IMDb Top 250
## Create a new dataframe IMDb_Top_250
```

```python
df_sorted = df.sort_values(by='imdb_score', ascending=False)
df_filtered = df_sorted[df_sorted['num_voted_users'] > 25000]
IMDb_Top_250 = df_filtered.head(250)

print("Top 250 movies with highest IMDb rating and num_voted_users > 25,000:")
print(IMDb_Top_250)
```

```python
IMDb_Top_250_sorted = IMDb_Top_250.sort_values(by='imdb_score', ascending=False)
IMDb_Top_250_filtered = IMDb_Top_250_sorted[IMDb_Top_250_sorted['num_voted_users'] > 25000]
IMDb_Top_250_filtered = IMDb_Top_250_filtered.head(250)
IMDb_Top_250_filtered['Rank'] = range(1, len(IMDb_Top_250_filtered) + 1)

print("Top 250 movies with highest IMDb rating and num_voted_users > 25,000 with Rank:")
print(IMDb_Top_250_filtered)
```

```python
Top_Foreign_Lang_Film = IMDb_Top_250[IMDb_Top_250['language'] != 'English']
print("Movies in IMDb_Top_250 DataFrame which are not in English language:")
print(Top_Foreign_Lang_Film)
```

```python
## Subtask 3.5: Find the best directors
# Group the dataframe using the director_name column.

grouped_df = df.groupby('director_name')
for director, group in grouped_df:
    print("Director:", director)
    print(group)
```

```python
#Top 10 - director


director_mean_scores = df.groupby('director_name')['imdb_score'].mean()
top10director = director_mean_scores.sort_values(ascending=False).head(10)

print("Top 10 directors with highest mean imdb_score:")
print(top10director)
```

```python
## Subtask 3.6: Find popular genres

split_genres = df['genres'].str.split('|', expand=True)
df['genre_1'] = split_genres[0]
df['genre_2'] = split_genres[1].fillna(split_genres[0])

print("DataFrame with genre_1 and genre_2 columns:")
print(df)
```

```python
#Group the dataframe using genre_1 as the primary column and genre_2 as the secondary column

grouped_df = df.groupby(['genre_1', 'genre_2'])
for group, movies in grouped_df:
    print("Genre 1:", group[0])
    print("Genre 2:", group[1])
    print("Movies:")
    print(movies)
    print()
```

```python
## Top 5 most popular combinations of genres

PopGenre = df.groupby(['genre_1', 'genre_2'])['gross'].mean().reset_index()
PopGenre = PopGenre.sort_values(by='gross', ascending=False).head(5)

print("Top 5 most popular combinations of genres by mean gross:")
print(PopGenre)
```

```python
## Subtask 3.7: Find the critic-favorite and audience-favorite actors
# creating three new dataframes here

Meryl_Streep = df[df['actor_1_name'] == 'Meryl Streep']
Leo_Caprio = df[df['actor_1_name'] == 'Leonardo DiCaprio']
Brad_Pitt = df[df['actor_1_name'] == 'Brad Pitt']

print("Meryl Streep's movies:")
print(Meryl_Streep)

print("\nLeonardo DiCaprio's movies:")
print(Leo_Caprio)

print("\nBrad Pitt's movies:")
print(Brad_Pitt)
```

```python
## Append the rows of all these dataframes and store them in a new dataframe named Combined

Meryl_Streep = pd.DataFrame({'actor_1_name': ['Meryl Streep', 'Meryl Streep'],
                             'movie_title': ['Movie1', 'Movie4']})

Leo_Caprio = pd.DataFrame({'actor_1_name': ['Leonardo DiCaprio', 'Leonardo DiCaprio'],
                           'movie_title': ['Movie2', 'Movie5']})

Brad_Pitt = pd.DataFrame({'actor_1_name': ['Brad Pitt'],
                          'movie_title': ['Movie3']})

# Concatenate the rows of Meryl_Streep, Leo_Caprio, and Brad_Pitt DataFrames
Combined = pd.concat([Meryl_Streep, Leo_Caprio, Brad_Pitt])

print("Combined DataFrame:")
print(Combined)
```

```python
##Group the combined dataframe using the actor_1_name column.

grouped_combined = Combined.groupby('actor_1_name')
for actor, group in grouped_combined:
    print("Actor:", actor)
    print("Movies:")
    print(group)
    print()
```

```python
## Find the mean of the num_critic_for_reviews and num_user_for_review and identify the actors which have the highest mean.

actor_means = Combined.groupby('actor_1_name')[['num_critic_for_reviews', 'num_user_for_reviews']].mean()
highest_mean_values = actor_means.max(axis=1)
actors_with_highest_mean = actor_means.index[actor_means.eq(highest_mean_values, axis=0).any(axis=1)].tolist()

print("Actors with the highest mean values:")
print(actors_with_highest_mean)
```
